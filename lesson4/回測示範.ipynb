```python
import jettool
import pandas
import numpy
tejtool = jettool.jet.engine()
tejtool.set_data()
tejtool.applied = True
```

```python
# 定義計算式，然後使用back_test函式即可回測，目前欠缺自動查詢函式內需要資料欄位(即quantjet網站js檔的功能)的程序

def calculate():

    tejtool.data['益本比排名'] = tejtool.sort_crossing('益本比',method='percentile',ascending=True)
    tejtool.data['股價淨值比排名'] = tejtool.sort_crossing('淨值市價比',method='percentile',ascending=True)
    tejtool.data['股利率排名'] = tejtool.sort_crossing('股利殖利率',method='percentile',ascending=True)
    
    tejtool.data['門檻條件'] = 0.70
    tejtool.data['益本比排名檢查'] = tejtool.check_above('益本比排名','門檻條件')
    tejtool.data['股價淨值比排名檢查'] = tejtool.check_above('股價淨值比排名','門檻條件')
    tejtool.data['股利率排名檢查'] = tejtool.check_above('股利率排名','門檻條件')
    
def evaluate():
    tejtool.data['購入檢查'] = tejtool.check_condition(check_type='and',conditions=['益本比排名檢查','股價淨值比排名檢查','股利率排名檢查'])
    tejtool.data['購入'] = tejtool.choose_setting('購入檢查',reset_list='01')
    tejtool.data['unit'] = tejtool.equal_pv()

# 另外必須輸入用來當作市價與報酬率的欄位名稱，如:roib_name='報酬率-Ln',closed_name ='收盤價(元)'
# 若不輸入，則使用此預設值
tejtool.back_test(back_length=700,back_interval=['2018-05-01','2020-04-01'],cash=1000000000,calculate=calculate,evaluate=evaluate)
```

```python
def calculate(tejtool):
    tejtool.data['報酬率']=tejtool.calculate_growthrate(check_index='收盤價(元)',window='1m',fix_date='01',method='geometric')
    tejtool.data['famamodel']=tejtool.make_famamacbethmodel(col_name='報酬率-Ln',check_index=['市值', '淨值市價比', '益本比', '股利殖利率', '動能(T-12~T-2)', '短期反轉(T-1)', '長期反轉(T-60~T-13)', '盈利能力(ROE)', '投資(資產成長率)', '違約距離(MERTON)', '違約距離(KMV)', '流動性', '個別風險', '產業集中度', '錯誤定價', '偏態係數', '研發密度'],window ='24m',alpha_rate = 5,reset_list = '01')
    tejtool.data['famascore']=tejtool.run_famascore(rank_above=100,class_count = 10) 
    tejtool.data['fama分組當月報酬']=tejtool.calculate_crossing(check_index='報酬率',window='1d',weight='市值',col_kind='mean')
def evaluate(tejtool):
    tejtool.data['fama分組判斷'] =tejtool.abnormal_selection(check_index='fama分組當月報酬',group_name='famascore',window='12m',keep='first')
    tejtool.data['購入']=tejtool.check_condition(conditions=['fama分組判斷'],check_type='and') 
    tejtool.data['unit'] = tejtool.equal_pv()
tejtool.back_test(back_interval='2019-3-29',cash=1000000000,calculate=calculate,evaluate=evaluate)
```

```python
fix_data_sample = tejtool.all_date_data[['coid','zdate','益本比','淨值市價比','股利殖利率']].merge(tejtool.hold_data[['zdate','coid','unit','roibNext']])
```

```python
fix_data_sample.to_csv('fix_data_sample.csv',index=False)
```

```python
fix_data_sample['5-roib'] = fix_data_sample['roibNext'].rolling(window=5).sum()
fix_data_sample['sdate'] = fix_data_sample['zdate'].astype(str).str[0:7].astype('datetime64')
fix_data_sample = fix_data_sample.drop_duplicates(subset=['coid','sdate'],keep='first')
fix_data_sample = fix_data_sample.dropna()
fix_data_sample
```

```python
fix_data_sample['購買'] = 0
fix_data_sample.loc[fix_data_sample['unit']>0,'購買'] = 1
x_filed = ['益本比','淨值市價比','股利殖利率']
y_field = '購買'
```

```python
fix_data_sample['購買'].describe()
```

```python
import xgboost
xgb_train = xgboost.DMatrix(fix_data_sample[x_filed].values, label=fix_data_sample[y_field].values)
xgb_params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
    }
bst = xgboost.train(xgb_params,
                    xgb_train,
                    num_boost_round=100,
                    )
```

```python
# 比較準確度，為什麼樣本較大，反而比較不準？
fix_data_sample['分類機率'] = bst.predict(xgb_train)
fix_data_sample['預測分組'] = 0
fix_data_sample.loc[fix_data_sample['分類機率']>0.5,'預測分組']  = 1
fix_data_sample['預測分組差異'] = fix_data_sample[y_field] - fix_data_sample['預測分組']
#印出猜中的比重
print([len(fix_data_sample[fix_data_sample[y_field]==0]),len(fix_data_sample[fix_data_sample[y_field]==1])])
print([len(fix_data_sample[fix_data_sample['預測分組']==0]),len(fix_data_sample[fix_data_sample['預測分組']==1])])
print(len(fix_data_sample[fix_data_sample['預測分組差異']==0])/len(fix_data_sample))
```

```python
# 計算各種true positive、false negative、false positive、true negative
print(len(fix_data_sample[(fix_data_sample['預測分組']==1)&(fix_data_sample[y_field]==1)])/len(fix_data_sample[fix_data_sample[y_field]==1]))
print(len(fix_data_sample[(fix_data_sample['預測分組']==0)&(fix_data_sample[y_field]==1)])/len(fix_data_sample[fix_data_sample[y_field]==1]))
print(len(fix_data_sample[(fix_data_sample['預測分組']==1)&(fix_data_sample[y_field]==0)])/len(fix_data_sample[fix_data_sample[y_field]==0]))
print(len(fix_data_sample[(fix_data_sample['預測分組']==0)&(fix_data_sample[y_field]==0)])/len(fix_data_sample[fix_data_sample[y_field]==0]))
```

```python
x_filed = ['益本比','淨值市價比','股利殖利率']
y_field = '5-roib'
features = fix_data_sample[x_filed].values
target = fix_data_sample[y_field].values
max_depth = 13
min_child_weight = 10
subsample = 0.5
colsample_bytree = 0.6
objective = 'reg:linear'
num_estimators = 1000
learning_rate = 0.3
eval_metric = 'logloss'
clf = xgboost.XGBRegressor(
                max_depth=max_depth,
                #min_child_weight=min_child_weight,
                #subsample=subsample,
                #colsample_bytree=colsample_bytree,
                #n_estimators=num_estimators,
                #learning_rate=learning_rate    
                objective=objective,
                eval_metric=eval_metric)
clf.fit(features, target)
```

```python
predict_roi = clf.predict(features)
fix_data_sample['xgbr_predict'] = predict_roi
fix_data_sample['xgbr_predict_diff'] = numpy.abs(fix_data_sample['xgbr_predict'] - fix_data_sample[y_field])/numpy.abs(fix_data_sample[y_field])
```
